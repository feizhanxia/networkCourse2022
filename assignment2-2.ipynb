```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = np.genfromtxt("Dataset for Homework2-2.txt",dtype=[int,int])
data_frame=pd.DataFrame(data)
G=nx.from_pandas_edgelist(data_frame,"f0","f1",create_using=nx.Graph())
```

### 平均聚类系数 Average Clustering Coefficient

```python
print("平均集聚系数：",nx.average_clustering(G))
```

### 度分布 及 平均度$<k>$

```python
d=dict(nx.degree(G))
print("Average Degree:%g"%(sum(d.values())/G.number_of_nodes()))
fig,ax1 = plt.subplots(figsize = (13,5),dpi = 300)
ax1.scatter(range(len(nx.degree_histogram(G))),np.array(nx.degree_histogram(G))/sum(nx.degree_histogram(G)),color="C4",s= 4) #在双对坐标轴上绘制度分布曲线
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_xlabel(r'$k$ (In Degree)')
ax1.set_ylabel(r'$p(k)$')
ax1.set_title('Degree Distribution')
ax1.text(1,0.001,r"$<k> = $%g"%(sum(d.values())/G.number_of_nodes()))
fig.savefig('2-2-1.png')
```

### 度-度相关性

```python
fig2,ax2 = plt.subplots(figsize = (8,6),dpi = 300)
im2 = ax2.imshow(nx.degree_mixing_matrix(G, mapping={x: x for x in range(max(deg for n, deg in G.degree) + 1)})[0:50,0:50],origin = 'lower',cmap = 'YlOrBr')  
fig2.colorbar(im2, ax=ax2, label='Correlation')
ax2.text(10,30,"Assortativity Coefficient:%g"%nx.degree_assortativity_coefficient(G))
ax2.set_xlabel(r"$k$")
ax2.set_ylabel(r"$k$")
fig2.savefig('2-2-2.png')
```
